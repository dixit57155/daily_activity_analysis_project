```python
# Extract weekday and create a new column 'day'
```

```python
# Convert 'ActivityDate' column to datetime if it's not already in datetime format
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])

# Now you can use .dt accessor to extract weekday
df['day'] = df['ActivityDate'].dt.weekday

df['day'] = df['ActivityDate'].dt.weekday
```

```python
df.dtypes
```

```python
weekday_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
df['day'] = df['day'].map(weekday_map)
```

```python
df
```

```python
df['Weekend'] = df['day'].apply(lambda x: 1 if x in ['Saturday', 'Sunday'] else 0)
```

```python
print(df[['ActivityDate', 'day', 'Weekend']])
```

# Visualize activity on weekends and weekdays

# Group the data by 'Weekend' and calculate the mean of activity minutes for each group

```python
activity_means = df.groupby('Weekend')[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes']].mean().reset_index()
```

```python
activity_means
```

```python
activity_means_melted = activity_means.melt(id_vars='Weekend', var_name='ActivityType', value_name='AverageMinutes')
```

```python
activity_means_melted
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='ActivityType', y='AverageMinutes', hue='Weekend', data=activity_means_melted)
plt.title('Average Activity Minutes on Weekdays vs. Weekends')
plt.xlabel('Activity Type')
plt.ylabel('Average Minutes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Correlation and Causation

*Correlation: a measure of the extent of interdependence between variables.

Causation: the relationship between cause and effect between two variables.

It is important to know the difference between these two. Correlation does not imply causation. Determining correlation is much simpler the determining causation as causation may require independent experimentation.*

```python
# import scipy library
from scipy import stats
```

## TotalSteps vs TotalDistance

```python
pearson_coef, p_value = stats.pearsonr(df['TotalSteps'], df['TotalDistance'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

- The correlation coefficient indicates a very strong positive linear relationship, while the p-value suggests the relationship is statistically significant.

## TotalSteps vs TrackerDistance

```python
pearson_coef, p_value = stats.pearsonr(df['TotalSteps'], df['TrackerDistance'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

- The correlation coefficient implies an extremely strong positive linear relationship, with a statistically significant association indicated by the p-value nearing zero.

## Calories vs TrackerDistance

```python
pearson_coef, p_value = stats.pearsonr(df['Calories'], df['TrackerDistance'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

- The correlation coefficient denotes a moderate positive linear relationship, with a highly significant association indicated by the extremely low p-value.

## Calories vs TotalSteps

```python
pearson_coef, p_value = stats.pearsonr(df['Calories'], df['TotalSteps'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

- The correlation coefficient suggests a moderate positive linear relationship, while the extremely low p-value indicates a highly significant association between the variables.
